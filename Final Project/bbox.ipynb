```python
bbox = []
for i,im in enumerate(train_x):
    img = cv2.cvtColor((im).astype(np.uint8), cv2.COLOR_BGR2RGB)
    r = cv2.selectROI("Image"+ str(i), img)
    bbox.append(r)
    print(bbox)
    
    if cv2.waitKey(1)==27:
        cv2.destroyWindow("Image"+ str(i))
```
